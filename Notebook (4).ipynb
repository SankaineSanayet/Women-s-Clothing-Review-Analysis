## 4. Data Preprocessing and Cleaning (Text-Specific)

  
After ensuring overall data quality, we now focus on cleaning the text data. We will process the “Review Text” (and optionally “Title”) fields using the following steps:
- **Lowercasing:** Convert text to lowercase.
- **Removing Noise:** Eliminate URLs, HTML tags, mentions, hashtags, punctuation, digits, extra spaces, and expand contractions.
- **Tokenization:** Split text into individual words (tokens).
- **Stopword Removal:** Remove common words that do not carry significant meaning.
- **Stemming or Lemmatization:** Reduce words to their base or dictionary forms.
- **N-gram Generation:** Generate bigrams to capture context and word order.

Each of these steps is implemented in separate cells.


 **a) Lowercasing:**  
  Convert all text to lowercase to ensure uniformity.


```python
# Lowercase the review text and store in a new column 'Clean_tweet'
df['Clean_text'] = df['Review Text'].str.lower()

df.head()
```


**b) Removing Noise:**  
  Clean the text by removing:
  - **URLs & HTML Tags:** Unnecessary links and markup.
  - **Mentions & Hashtags:** Social media markers (if present) unless contextually relevant.
  - **Punctuation, Special Characters & Digits:** Remove non-informative symbols.
  - **Extra Whitespace:** Clean up unnecessary spaces.
  - **Contractions:** Expand shortened forms (e.g., "can't" to "cannot").


```python
# Function to clean text
import re
import string
import contractions 

def clean_text(text):
    """
    Remove unwanted characters and noise:
    - Expand contractions.
    - Remove URLs, HTML tags, mentions, hashtags, punctuation, and digits.
    - Remove extra spaces.
    """
    if not isinstance(text, str):
        return text  # Skip non-string values
    text = contractions.fix(text)  # Expand contractions (e.g., "can't" -> "cannot")
    text = re.sub(r'http\S+|www\S+|https\S+', '', text)  # Remove URLs
    text = re.sub(r'<.*?>', '', text)  # Remove HTML tags
    text = re.sub(r'@\w+|#\w+', '', text)  # Remove mentions and hashtags
    text = text.translate(str.maketrans('', '', string.punctuation))  # Remove punctuation
    text = re.sub(r'\d+', '', text)  # Remove digits
    text = re.sub(r'\s+', ' ', text).strip()  # Remove extra spaces
    return text

# Apply noise removal to the lowercased text and store result in 'clean_text'
df['Clean_text'] = df['Clean_text'].apply(clean_text)

pd.set_option("display.max_colwidth", 400)
df[['Review Text', 'Clean_text']]
```

**c) Tokenization:**  
  Split the cleaned text into individual words (tokens).



```python
import nltk

# Tokenize the cleaned text into individual words and store the tokens in 'tokens'
df['Clean_text'] = df['Clean_text'].apply(nltk.word_tokenize)

# Display a sample of the tokenized text
df[['Review Text', 'Clean_text']].head()
```

 **d) Stopword Removal:**  
  Remove common stopwords (e.g., "the", "is") to focus on more meaningful words.

```python
from nltk.corpus import stopwords
stop_words = set(stopwords.words('english'))

# Remove stopwords from the token list for each review; store the result in 'tokens_no_stop'
df['Clean_text'] = df['Clean_text'].apply(lambda tokens: [word for word in tokens if word not in stop_words])

# Display tokens before and after stopword removal for comparison
df[['Review Text', 'Clean_text']].head()
```

**e) Lemmatization:**  
  Reduce tokens to their base forms for standardization.

```python
from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()

# Apply lemmatization to the tokens (after stopword removal) and store the result in 'clean_tokens'
df['Clean_text'] = df['Clean_text'].apply(lambda tokens: [lemmatizer.lemmatize(word) for word in tokens])

# Display tokens before and after lemmatization for verification
df[['Review Text', 'Clean_text']].head()
```

### Word cloud of Positive and Negative Sentiments

```python
import matplotlib.pyplot as plt
from wordcloud import WordCloud

def generate_sentiment_wordclouds(df):
    """
    Generates word clouds for both positive and negative sentiment reviews.

    Parameters:
    df (DataFrame): Pandas DataFrame containing 'Sentiment' and 'Clean_text' columns.

    Returns:
    Displays two word clouds: one for positive sentiment and one for negative sentiment.
    """
    # Filter text based on sentiment
    positive_reviews = df[df["Sentiment"] == "Positive"]["Clean_text"]
    negative_reviews = df[df["Sentiment"] == "Negative"]["Clean_text"]

    # Convert lists to strings if necessary
    positive_reviews = positive_reviews.apply(lambda x: " ".join(x) if isinstance(x, list) else x)
    negative_reviews = negative_reviews.apply(lambda x: " ".join(x) if isinstance(x, list) else x)

    # Convert reviews into a single text string
    positive_text = " ".join(positive_reviews)
    negative_text = " ".join(negative_reviews)

    # Generate Word Clouds
    positive_wordcloud = WordCloud(width=800, height=400, background_color="white", colormap="Blues").generate(positive_text)
    negative_wordcloud = WordCloud(width=800, height=400, background_color="black", colormap="Reds").generate(negative_text)

    # Plot both word clouds side by side
    plt.figure(figsize=(14, 6))

    # Positive Word Cloud
    plt.subplot(1, 2, 1)
    plt.imshow(positive_wordcloud, interpolation="bilinear")
    plt.axis("off")
    plt.title("Word Cloud - Positive Sentiment", fontsize=14)

    # Negative Word Cloud
    plt.subplot(1, 2, 2)
    plt.imshow(negative_wordcloud, interpolation="bilinear")
    plt.axis("off")
    plt.title("Word Cloud - Negative Sentiment", fontsize=14)

    # Show the plots
    plt.show()

# Call the function
generate_sentiment_wordclouds(df)
```

### Most Common Bigrams for both Positive and Negative Sentiments 



```python
import pandas as pd
from collections import Counter
import matplotlib.pyplot as plt
import seaborn as sns

def plot_common_bigrams(df, sentiment="Positive", top_n=20):
    """
    Plots the most common bigrams for a given sentiment.

    Parameters:
    df (pd.DataFrame): DataFrame containing 'Clean_text' and 'Sentiment' columns.
    sentiment (str): "Positive" or "Negative".
    top_n (int): Number of top bigrams to display.

    Returns:
    None (displays a bar chart)
    """

    # Check if dataframe is empty
    if df.empty:
        print("The dataframe is empty!")
        return

    # Filter by sentiment
    sentiment_df = df[df["Sentiment"] == sentiment]

    # Ensure Clean_text is a list of tokens
    all_bigrams = [
        tuple(text[i:i+2]) 
        for text in sentiment_df["Clean_text"] 
        if isinstance(text, list) 
        for i in range(len(text)-1)
    ]
    
    # Count bigram frequencies
    bigram_counts = Counter(all_bigrams)

    # Get top N bigrams
    most_common_bigrams = bigram_counts.most_common(top_n)

    # If no bigrams exist, return early
    if not most_common_bigrams:
        print(f"No bigrams found for {sentiment} sentiment.")
        return

    bigrams, counts = zip(*most_common_bigrams)

    # Convert bigrams to readable format
    bigram_labels = [" ".join(bigram) for bigram in bigrams]

    # Debugging Print Statements
    print("Bigrams:", bigram_labels)
    print("Counts:", counts)

    # Plot bar chart
    plt.figure(figsize=(12, 6))
    sns.barplot(x=list(counts), y=bigram_labels, color="skyblue")  
    plt.xlabel("Frequency")
    plt.ylabel("Bigrams")
    plt.title(f"Top {top_n} Most Common Bigrams for {sentiment} Sentiment")
    plt.show()

# Example Usage:
plot_common_bigrams(df, sentiment="Positive", top_n=20)
plot_common_bigrams(df, sentiment="Negative", top_n=20)
```

## 5. Feature Engineering and Representation

  
To convert our text into a format that machine learning models can process, we transform the final cleaned review text into numerical features. This process consists of two parts:

### Text Vectorization:
- **Objective:** Convert the preprocessed text into numerical features.
- **Method:** We use the TF-IDF technique, which assigns weights to words based on their frequency in each review relative to the entire dataset.
- **Inclusion of Unigrams and N-grams:** Both unigrams and bigrams are included to capture individual words and contextual word pairs.

```python
from sklearn.feature_extraction.text import TfidfVectorizer

# Initialize TF-IDF Vectorizer
tfidf_vectorizer = TfidfVectorizer(ngram_range=(1,2), max_features=5000)  # Unigrams & Bigrams, 5000 features

# Apply TF-IDF on "final_clean" column
X = tfidf_vectorizer.fit_transform(df["final_clean"])

# Print the shape of the TF-IDF feature matrix to verify dimensions
print("TF-IDF feature matrix shape:", X.shape)

# Define target variable (Recommended)
y = df["Recommended IND"]
```

## Feature Selection / Dimensionality Reduction

  
The TF-IDF vectorization produces a high-dimensional feature matrix, which can lead to inefficiencies and overfitting. To address this, we can apply dimensionality reduction techniques.  
- **Method:** We use TruncatedSVD (which is similar to PCA but designed for sparse data) to reduce the number of features while retaining the most important information.

- **Non-Negative Matrix Factorization (NMF)** ensures all values remain non-negative (perfect for MultinomialNB)

- **Objective:** Reduce the dimensionality of our feature matrix to improve computational efficiency and potentially enhance model performance.


```python
from sklearn.decomposition import TruncatedSVD

# Initialize Truncated SVD
n_components = 100  # Number of reduced dimensions (adjustable)
svd = TruncatedSVD(n_components=n_components, random_state=42)

# Apply SVD on TF-IDF features
X_svd = svd.fit_transform(X)

# Print new shape
print("Original Shape:", X.shape, " → Reduced Shape:", X_svd.shape)
```

## 6. Model Selection and Training

  
With our data preprocessed and converted into numerical features, we now move on to model selection and training. Our process includes:

1. **Baseline Model:**  
   - We start with a simple model (e.g., Logistic Regression) to establish a performance benchmark.
   - This baseline model helps us understand the minimum performance achievable with minimal complexity.
   - We'll use Smote to handle Class imbalance in the the Target Variable


```python
from imblearn.over_sampling import SMOTE
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import train_test_split

# Split dataset into 80% training and 20% testing
X_train, X_test, y_train, y_test = train_test_split(
    X_svd, y, test_size=0.2, random_state=42, stratify=y
)

# Print dataset shapes before SMOTE
print("Before SMOTE:")
print("Training Data Shape:", X_train.shape)
print("Training Labels Distribution:\n", y_train.value_counts())

# Apply SMOTE to balance the classes in the training data
smote = SMOTE(random_state=42)
X_train_sm, y_train_sm = smote.fit_resample(X_train, y_train)

# Print dataset shapes after SMOTE
print("\nAfter SMOTE:")
print("Resampled Training Data Shape:", X_train_sm.shape)
print("Resampled Training Labels Distribution:\n", y_train_sm.value_counts())

# Initialize Logistic Regression model
logreg = LogisticRegression(random_state=42, max_iter=500)

# Train on SMOTE-balanced data
logreg.fit(X_train_sm, y_train_sm)

# Predict on test data
y_pred = logreg.predict(X_test)

# Evaluate model
accuracy = accuracy_score(y_test, y_pred)
print("\nBaseline Logistic Regression Accuracy with SMOTE:", accuracy)
print("\nClassification Report:\n", classification_report(y_test, y_pred))
```

## Hyperparameter Tuning & Cross-Validation

## **Multinomial Naive Bayes** 

### **Steps** 

**1. Feature Transformation Using NMF**

- **Non-Negative Matrix Factorization (NMF)** reduces the dimensionality of the TF-IDF feature space while ensuring all values are non-negative.

- This is crucial because **MultinomialNB requires non-negative inputs** (negative values would break probability calculations).

**2. Handling Class Imbalance with SMOTE**

- **Synthetic Minority Over-sampling Technique (SMOTE)** is applied to the training data to generate synthetic samples for the minority class.

- This helps improve model performance, especially when one class is significantly underrepresented.

**3. Model Training and Evaluation**

- **The MultinomialNB classifier** is trained on the **SMOTE-balanced dataset**.

- After training, predictions are made on the test set.

- The accuracy and classification report (precision, recall, F1-score) are computed to assess the model’s performance.

```python
from sklearn.decomposition import NMF
from imblearn.over_sampling import SMOTE
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, classification_report
from sklearn.model_selection import train_test_split

# Initialize NMF (ensures all values are non-negative)
n_components = 100  # Number of reduced dimensions (adjustable)
nmf = NMF(n_components=n_components, random_state=42)

# Apply NMF on TF-IDF features
X_nmf = nmf.fit_transform(X)  # Ensures non-negative values

# Split dataset into 80% training and 20% testing
X_train_nb, X_test_nb, y_train_nb, y_test_nb = train_test_split(
    X_nmf, y, test_size=0.2, random_state=42, stratify=y
)

# Apply SMOTE to balance the classes in the training data
smote = SMOTE(random_state=42)
X_train_resampled, y_train_resampled = smote.fit_resample(X_train_nb, y_train_nb)

# Initialize MultinomialNB model
nb = MultinomialNB()

# Train on SMOTE-balanced data
nb.fit(X_train_resampled, y_train_resampled)

# Predict on test data
y_pred_nb = nb.predict(X_test_nb)

# Evaluate model
accuracy = accuracy_score(y_test_nb, y_pred_nb)
print("\nMultinomialNB Accuracy with NMF & SMOTE:", accuracy)
print("\nClassification Report:\n", classification_report(y_test_nb, y_pred_nb))
```
